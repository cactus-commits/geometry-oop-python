```python
from circle import Circle

# 1. Create a circle
c1 = Circle(x=0, y=0, radius=5)
print(c1)

# 2. Check area and perimeter 
print(f"Area: {c1.area}")
print(f"Perimeter: {c1.perimeter}")

# 3. Check if it's a unit circle
print(f"Is unit circle: {c1.is_unit_circle()}")

# 4. Test translater method (inherit from Shape)
c1.translate(2,4)
print(f"The new coordinates for c1 are ({c1.x}, {c1.y})")

# 5. Comparison
c2 = Circle(radius=3)
print(f"c1 > c2: {c1 > c2}")
print(f"c1 == c2: {c1 == c2}")
```

```python
from rectangle import Rectangle

# 1. Create a triangle
r1 = Rectangle(x=0, y=0, width=3, height=4)
print(r1)

# 2. Check area and perimeter
print(f"Area: {r1.area}")
print(f"Perimeter: {r1.perimeter}")

# 3. Check if it's a square
print(f"Is square: {r1.is_square()}")

# 4. Test translater method (inherit from Shape)
r1.translate(2,4)
print(f"The new coordinates for r1 are ({r1.x}, {r1.y})")

# 5. Comparison
r2 = Rectangle(width=5, height=8)
print(f"r1 < r2: {r1 < r2}")
print(f"c1 == c2: {r1 == r2}")
```

```python
from sphere import Sphere

s = Sphere(x=0, y=0, z=0, radius=5)
print(s)

print(f"Surface area: {s.surface_area}")
print(f"Volume: {s.volume}")
print(f"Is spehere unit sphere: {s.is_unit_sphere()}")
```
